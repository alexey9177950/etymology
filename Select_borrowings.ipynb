```python
approved_relations = set([
    'borrowed_from', # заимствовано -- однозначно подходит
    'derived_from', # норм
    'inherited_from', # подходит
    'abbreviation_of', # etc, etc
    'clipping_of', # скорее всего это будет из английского, но оставим
    'compound_of', # для слов типа watchlist. надо обрабатывать отдельно (+1 уровень вложенности),
    'calque_of', # калька -- хорошо
    'blend_of', # как compound но обрезаны конец и начало соединяемых слов
    'unadapted_borrowing_from',
    'learned_borrowing_from',
    'orthographic_borrowing_from',
    'named_after',
    'semantic_loan_of',
    'phono-semantic_matching_of',
    'initialism_of',
])


def select_borrowings(langs_from, langs_to, filter_uppercase=True):
    data = []
    header = None
    n_colums = None
    cnt_lines = 0
    broken_lines_cnt = 0


    # https://github.com/droher/etymology-db
    for line in open("/home/alexey/Downloads/etymology.csv"):
        cnt_lines += 1
    
        if header is None:
            header = line.split(',')
            n_columns = len(header)
            continue

        tokens = line.split(',')
        if len(tokens) < n_columns:
            broken_lines_cnt += 1
            continue

        lang1 = tokens[1]    
        term1 = tokens[2]
        reltype = tokens[3]
        lang2 = tokens[5]
        term2 = tokens[6]
    
    
        if lang1 == '' or term1 == '' or lang2 == '' or reltype == '':
            broken_lines_cnt += 1
            continue
        
        if lang1 in langs_to and lang2 in langs_from and reltype in approved_relations and \
        (not filter_uppercase or term1[0].islower()):
            data.append(((lang1, term1), reltype, (lang2, term2)))

    return data


def print_borrowings(langs_from, langs_to, limit=30, filter_uppercase=True):
    data = sorted(
        select_borrowings(langs_from, langs_to),
        key = lambda x : x[0][1]
    )
    print(len(data), "items")
    
    if limit is None:
        iterlist = zip(data, range(len(data)))
    else:
        iterlist = zip(data, range(limit))
    for line, _ in iterlist:
        print(line)
    if limit is not None and len(data) > limit:
        print("...")
```
