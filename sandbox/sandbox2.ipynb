```python

data = []
header = None
n_colums = None
cnt_lines = 0
broken_lines_cnt = 0

selected_lines_1 = []
selected_lines_2 = []

# https://github.com/droher/etymology-db
for line in open("/home/alexey/Downloads/etymology.csv"):
    cnt_lines += 1
    
    if header is None:
        header = line.split(',')
        n_columns = len(header)
        print("Header:")
        print(line)
        continue

    tokens = line.split(',')
    if len(tokens) < n_columns:
        broken_lines_cnt += 1
        continue

    lang1 = tokens[1]    
    term1 = tokens[2]
    reltype = tokens[3]
    lang2 = tokens[5]
    term2 = tokens[6]
    
    
    if lang1 == '' or term1 == '' or lang2 == '' or reltype == '':
        broken_lines_cnt += 1
        continue
    
    to_append = (lang1, term1, reltype, lang2, term2)
    
    if lang1 in ["Russian", "English"] and lang2 in ['Yiddish', 'Hebrew']:
        selected_lines_1.append(to_append)
    if lang1 == "English" and lang2 == "Russian":
        selected_lines_2.append(to_append)

print(cnt_lines, broken_lines_cnt, len(selected_lines_1), len(selected_lines_2))
```

```python
for line in selected_lines_1:
    print(line)
```

```python
approved_relations = set([
    'borrowed_from', # заимствовано -- однозначно подходит
    'derived_from', # норм
    'inherited_from', # подходит
    'abbreviation_of', # etc, etc
    'clipping_of', # скорее всего это будет из английского, но оставим
    'compound_of', # для слов типа watchlist. надо обрабатывать отдельно (+1 уровень вложенности),
    'calque_of', # калька -- хорошо
    'blend_of', # как compound но обрезаны конец и начало соединяемых слов
    'unadapted_borrowing_from',
    'learned_borrowing_from',
    'orthographic_borrowing_from',
    'named_after',
    'semantic_loan_of',
    'phono-semantic_matching_of',
    'initialism_of',
])
```

```python
selected_lines_1_direct = [i for i in selected_lines_1 if i[2] in approved_relations]
print(len(selected_lines_1_direct))
print()

for line in selected_lines_1_direct:
    print(line)
```

```python
selected_lines_2_direct = [i for i in selected_lines_2 if i[2] in approved_relations]
print(len(selected_lines_2_direct))
print()

for line in selected_lines_2_direct:
    print(line)
```
